```python
print("mean")
mean_salary = df.groupby("experience_level")["salary"].mean().reset_index()
print(mean_salary)
print("     ")
print("median")
print(df.groupby("experience_level")["salary"].median())
```

```python
plt.bar(mean_salary['experience_level'], mean_salary['salary'], color='skyblue')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary')

plt.show()
```
