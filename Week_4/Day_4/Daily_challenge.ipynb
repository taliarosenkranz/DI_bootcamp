```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

Instructions

    - Download and import the Data Science Job Salary dataset.
    - Normalize the ‘salary’ using Min-Max normalization.
    - Implement dimensionality reduction on the dataset.
    - Aggregate data by ‘experience_level’, calculating average and median salaries.


```python
df = pd.read_csv("datascience_salaries.csv")
```

```python
df
```

```python

scaler = MinMaxScaler()
df['salary_normalized'] = scaler.fit_transform(df[['salary']])
```

```python
df
```

```python
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
```

```python
df = pd.get_dummies(df, columns = cat_cols)
```

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df)
```

```python
reduced_data
```

```python
df
```

```python
mean_salary_by_experience = df.groupby('experience_level')['salary'].mean()
mean_salary_by_experience
```

```python
median_salary_by_experience = df.groupby('experience_level')['salary'].median()
median_salary_by_experience
```
