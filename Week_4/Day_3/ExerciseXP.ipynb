## Class Challenge

```python
import pandas as pd
```

```python
titanic_data = pd.read_csv("titanic.csv")
```

```python
titanic_data["Fare"]
```

```python
# Identify and handle outliers in 'Fare' column
Q1 = titanic_data['Fare'].quantile(0.25)
Q3 = titanic_data['Fare'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#removing outliers by only keeping values that are over the lower bound and hier than upper bound
titanic_data = titanic_data[(titanic_data['Fare'] >= lower_bound) & (titanic_data['Fare'] <= upper_bound)]


```

```python
print("Q1:",Q1)
print("Q3:", Q3)
print("IQR_", IQR)
print("lower_bound: ", lower_bound)
print("upper_bound: ", upper_bound)
```
