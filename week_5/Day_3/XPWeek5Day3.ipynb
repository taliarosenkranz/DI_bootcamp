```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from plotnine import ggplot, aes, geom_point, labs, geom_bar,theme
import plotly.graph_objects as go
import plotly.express as px
```

```python
#Create a pandas DataFrame with two columns, x and y, each containing 20 random numbers.
x =random.sample(range(20, 101), 20)
y = random.sample(range(20, 101), 20)
df = pd.DataFrame({'x':x, 'y':y})
```

```python
df
```

1. Create a basic scatter plot using Plotnine to visualize the relationship between two variables.

```python
# Create the scatter plot
scatter_plot = (
    ggplot(df, aes(x='x', y='y')) +
    geom_point() +
    labs(x='X-Value', y='Y-Value', title='Basic Scatter Plot')
)


print(scatter_plot)
```

2.  Use Plotnine to create a bar plot showing the frequency of categories in a dataset.

```python
bar_plot = (
    ggplot(df, aes(x=x, y=y)) +
    geom_bar(stat='identity') +
    labs(x='Category', y='Value', title='Category Frequency') +
    theme(figure_size=(6, 4)))

print(bar_plot)
```

3.
Create a pandas DataFrame with three columns: x, y1, and y2, where x is a sequence of numbers and y1, y2 are two different mathematical functions of x (e.g., sin(x) and cos(x)).

```python
x = random.sample(range(10,100),50)
y1 = np.sin(x)
y2 = np.cos(x)
df = pd.DataFrame({"x":x, "y1":y1, "y2":y2})
df.head()
```

3.2. Plot y1 and y2 against x in the same plot, using different colors for each line.
Add a legend to distinguish between the two lines, and label axes as X-Value, Y-Value, and add a title Comparison of Two Functions.

```python
plt.plot(df.x,df.y1, label='y1')
plt.plot(df.x,df.y2,label='y2')
plt.legend()
plt.title("y1 vs y2 over x")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

4.
Create a basic line chart using Plotly to visualize the trend of sales over 12 months. Assume the following monthly sales data:

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [200, 210, 250, 300, 320, 340, 360, 380, 390, 400, 420, 440]

# Create a line plot
fig = go.Figure(data=go.Scatter(x=months, y=sales, mode='lines'))

# Update layout
fig.update_layout(title='Line Plot',
                  xaxis_title='month Axis',
                  yaxis_title='sales Axis')

# Show plot
fig.show()
```

5.Using Plotly, create a bar chart to represent the number of books sold in different genres. Use the following data:

```python
genres = ['Fiction', 'Non-Fiction', 'Sci-Fi', 'Mystery', 'Biography']
books_sold = [120, 150, 90, 80, 130]

fig = go.Figure(data=go.Bar(x=genres, y=books_sold))

# Update layout
fig.update_layout(title='Bar Chart',
                  xaxis_title='genres',
                  yaxis_title='books sold')

# Show plot
fig.show()
```

6.Create a scatter plot to visualize the relationship between Age, Salary, and Years of Experience in a dataset. Assume the following data:

```python
data = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Years of Experience': [2, 5, 7, 10, 12]
})

fig = px.scatter(data, x='Age', y='Salary', size='Years of Experience', color='Years of Experience',
                 title='Scatter Plot with Size and Color Dimensions',
                 labels={'Age': 'Age', 'Salary': 'Salary', 'Years of Experience': 'Years of Experience'})

fig.show()
#sns.scatterplot(x='Years of Experience', y="Salary",data=data, hue="Age")
```
