üåü Exercise 2 : Creating A Line Plot For Temperature Variation
Instructions
Create a simple line plot using Matplotlib that represents the temperature variations over a week.
Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]) and label the x-axis as ‚ÄúDay‚Äù and the y-axis as ‚ÄúTemperature (¬∞F)‚Äù.


```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
import datetime as dt
```

```python
dates = pd.date_range('2024-03-01', periods=7) #7 days of the week
temperatures =  [72, 74, 76, 80, 82, 78, 75]
data = {"Date": dates,"Temperature":temperatures}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
plt.figure()
plt.plot(dates, temperatures)
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.xticks(rotation = 45)
plt.title("Dayly temperature variations")
plt.show()
```

üåü Exercise 4 : Data Visualisation
Instructions
Using the dataset sales_data.csv

- Calculate the total quantity of products sold by the company.
- Identify the category that had the highest revenue and calculate how much revenue it generated.
- Determine the average revenue per sale in the dataset.
- Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).


```python
df = pd.read_csv("sales_data.csv")
```

```python
df.head()
```

```python
total_quantity = df["quantity"].sum()
print("total quantity of products sold: ", total_quantity)
```

```python
highest_revenue = df.groupby("product")["price"].max()
print(highest_revenue)
total_revenue_laptop = df[df["product"] == 'Laptop']['revenue'].sum()
print("total revenue : ", total_revenue_laptop)
```

```python
average_revenue_per_sales = df["revenue"].mean()
print("average revenue per sale: ", round(average_revenue_per_sales, 3))
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
total_rev_per_month = df.groupby(df["date"].dt.month).sum()
```

```python
total_rev_per_quarter = df.groupby(df['date'].dt.quarter).sum()
```

```python
total_rev_per_quarter
```

```python
quartiles = ["Q1", "Q2", "Q3", "Q4"]
```

```python
plt.figure()
plt.bar(quartiles, total_rev_per_quarter.revenue)
plt.xlabel("Yearly Quarters")
plt.ylabel("Total Revenue")
plt.title("total revenue generated in each quarter of the year")
```
