```python
df_regions_gdp_lx = df.groupby("Country or region")[["GDP per capita", "Healthy life expectancy"]].mean().reset_index()
```

```python
top_10_gdp = df_regions_gdp_lx.sort_values("GDP per capita", ascending = False).head(10)
top_10_health = df_regions_gdp_lx.sort_values("Healthy life expectancy", ascending = False).head(10)
```

```python
fig, ([ax1, ax2]) = plt.subplots(2,1)
ax1.scatter(top_10_gdp["GDP per capita"], top_10_gdp["Country or region"])
ax1.set_title('top GDP per region')

ax2.scatter(top_10_health["Healthy life expectancy"], top_10_health["Country or region"])
ax2.set_title('top healthy life per region')
```

```python
x = ['Mexico', 'Israel', 'Germany', 'Spain', 'United Arab Emirates']
```

```python
df_selected = df[df["Country or region"].isin(x)]
```

```python
x = df_selected["Country or region"]
```

```python
y1 = df_selected["GDP per capita"]
y2 = df_selected["Healthy life expectancy"]
```

```python
df_selected["GDP per capita"]
```

```python
df_selected["Healthy life expectancy"]
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))
# Bar chart for GDP per capita
ax1.bar(x, y1, color='skyblue', label='GDP per capita')
ax1.set_xlabel('Country')
ax1.set_ylabel('GDP per capita ($)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Create a twin Axes sharing the xaxis for Healthy Life Expectancy
ax2 = ax1.twinx()
# Line chart for Healthy Life Expectancy
ax2.plot(x, y2, color='darkgreen', marker='o', label='Healthy Life Expectancy')
ax2.set_ylabel('Healthy Life Expectancy (Years)', color='darkgreen')
ax2.tick_params(axis='y', labelcolor='darkgreen')

# Title and customizations
fig.suptitle('How Economic Strength Relates to Health Outcomes in Different Regions')

# Fixing the legend - combining legends from both ax1 and ax2
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc="upper left", bbox_to_anchor=(0.1,0.9))

plt.show()
```
